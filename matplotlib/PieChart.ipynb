```python
import matplotlib.pyplot as plt 

brands = ["oneplus", "apple", "samsung", "nokia", "redmi"]
x = [22, 35, 30, 3, 10]
c = ["red", "violet", "blue", "magenta", "orange"]
ex = [0,0,0,0,0.1]

plt.pie(x, labels= brands, colors= c, explode=ex, shadow= True, autopct="%.2f", startangle=90)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_excel("expense3.xlsx")
df = pd.DataFrame(data)
grouped_by = df.groupby("Payment Mode")["Amount"].sum()
print(grouped_by)
ex = [0,0.1,0]
plt.pie(grouped_by.values, labels=grouped_by.index,  autopct="%.2f", startangle=45,explode=ex, shadow=True)
plt.show()
```
