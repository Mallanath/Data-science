# Box plot

Min (lower fence)
Q1 -> 25% =25/100 * (8 + 1) 

Median 

Q3 -> 75% = 7/100 * 9 
Max (upper fence)

Min/LF = Q1 - 1.5(IQR)
Max/UF = Q3 + 1.5(IQR)
IQR = Q3 - Q1

[1,3,4,7,12,2,8,9,24] = 70 => median = 70/10 = 7

Q1 = 25/100 * 10 = 2.5

Q2 = 75/100 * 10 = 7.5

IQR = 7.5 - 2.5 = 5

LF = 2.5 - 1.5(5) = -5
UF = 7.5 + 1.5(5) = 15

LF = -5   min = 1

UF = 15   max = 12

Q1 = 2.5  median = 7.5

Q2 = 7.5

IQR = 5





```python
import matplotlib.pyplot as plt

l = [25, 46, 57, 25, 45, 13, 29, 39, 46, 99, 22, 12, 14, 16, 55, 6, 42, 52, 55]

plt.boxplot(l)
plt.show()
```

```python
import matplotlib.pyplot as plt

l = [25, 46, 57, 25, 45, 13, 29, 39, 46, 99, 22, 12, 14, 16, 55, 6, 42, 52, 55]
l1 = [1,3,4,7,12,2,8,9,24]

plot_values = [l,l1]
plt.boxplot(plot_values)
plt.show()
```
